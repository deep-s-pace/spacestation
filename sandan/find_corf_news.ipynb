```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#회사이름 리스트가 담긴 csv를 읽어오자
result = []
for i in range(1325, 1387):
    csv = pd.read_csv('./all_company_list.csv')
    company = csv['company']
    company = company[i]
    print(company)
    
#     df = 'df_{}'.format(i) 
#     df = pd.DataFrame(columns=['title', 'source', 'head', 'url'])
    
    # 반복1: 기사번호를 변경시키면서 데이터 수집을 반복하기
    # 1 ~ 100까지 10단위로 반복(1, 11, ..., 91)
    raw = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}&start=".format(company),
                            headers={'User-Agent':'Mozilla/5.0'})
    html = BeautifulSoup(raw.text, "html.parser")

    articles = html.select("ul.list_news > li")

    # 반복2: 기사에 대해서 반복하면 세부 정보 수집하기                
    # 리스트를 사용한 반복문으로 모든 기사에 대해서 제목/언론사 출력
        
    for ar in articles:
                
        title = ar.select_one("a.news_tit").text
        source = ar.select_one("a.info.press").text
        head = ar.select_one("div.dsc_wrap").text
        url = ar.select_one('a.news_tit')["href"]

        print("title = ",title, "source = ", source, "head = ", head, "url = ", url)
        li = [company, title, source, head, url]
        result.append(li)

#         df.loc[len(df)]=[title, source, head, url]
    
#         news_df.append({'title':title, 'source':source, 'head':head, 'url':url}, ignore_index=True)
```

```python
import requests
from pandas import DataFrame
from bs4 import BeautifulSoup
import re
from datetime import datetime
import os
```

```python
date = str(datetime.now()) 
date = date[:date.rfind(':')].replace(' ', '_') 
date = date.replace(':','시') + '분' 
```

```python
query = input('(검색 키워드를 입력하세요 : ') 
query = query.replace(' ', '+') 

news_num = int(input('총 필요한 뉴스기사 수를 입력해주세요(숫자만 입력) : ')) 
```

```python
news_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'

req = requests.get(news_url.format(query))
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
news_dict = {} 
idx = 0 
cur_page = 1
```

```python
print()
print('크롤링 중...')

while idx < news_num:
### 네이버 뉴스 웹페이지 구성이 바뀌어 태그명, class 속성 값 등을 수정함(20210126) ###
    
    table = soup.find('ul',{'class' : 'list_news'})
    li_list = table.find_all('li', {'id': re.compile('sp_nws.*')})
    area_list = [li.find('div', {'class' : 'news_area'}) for li in li_list]
    a_list = [area.find('a', {'class' : 'news_tit'}) for area in area_list]
    
    for n in a_list[:min(len(a_list), news_num-idx)]:
        news_dict[idx] = {'title' : n.get('title'),
                          'url' : n.get('href') }
        idx += 1

    cur_page += 1
    
    pages = soup.find('div', {'class' : 'sc_page_inner'})
    next_page_url = [p for p in pages.find_all('a') if p.text == str(cur_page)][0].get('href')
    
    req = requests.get('https://search.naver.com/search.naver' + next_page_url)
    soup = BeautifulSoup(req.text, 'html.parser')
```

```python
print('크롤링 완료')

print('데이터프레임 변환')
news_df = DataFrame(news_dict).T

folder_path = os.getcwd()
xlsx_file_name = '네이버뉴스_{}_{}.xlsx'.format(query, date)

news_df.to_excel(xlsx_file_name)

print('엑셀 저장 완료 | 경로 : {}\\{}'.format(folder_path, xlsx_file_name))
os.startfile(folder_path)
```
