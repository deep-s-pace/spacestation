## 매출증가

```python
import matplotlib.pyplot as plot
import numpy as np
import math
import pandas as pd
import openpyxl

newData = data[["Q8","Q9","Q10","Q11","Q12","Q13","Q14","Q15","Q16","Q17","Q18"]]

#속성(변수) 선택
x = newData[['Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18']]
y = newData[['Q8']]

from sklearn.svm import SVR
import seaborn as sns

for i in (0.1, 1, 10, 100, 1000):
    for j in (0.1, 1, 10, 100, 1000):
        model = SVR(kernel = 'poly', C=i, gamma = j)
        model.fit(x, y)
        relation_square = model.score(x, y)
        print('결정계수 R : ', relation_square)
y_p = model.predict(x)

ax1 = sns.distplot(y, hist = False, label = 'y실제')
ax2 = sns.distplot(y_p, hist = False, label = 'y예측')
plot.show()
plot.close()
```

## 비용절감

```python
import matplotlib.pyplot as plot
import numpy as np
import math
import pandas as pd
import openpyxl

newData = data[["Q8","Q9","Q10","Q11","Q12","Q13","Q14","Q15","Q16","Q17","Q18"]]

#속성(변수) 선택
x = newData[['Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18']]
y = newData[['Q10']]

from sklearn.svm import SVR
import seaborn as sns

for i in (0.1, 1, 10, 100, 1000):
    for j in (0.1, 1, 10, 100, 1000):
        model = SVR(kernel = 'poly', C=i, gamma = j)
        model.fit(x, y)
        relation_square = model.score(x, y)
        print('결정계수 R : ', relation_square)
y_p = model.predict(x)

ax1 = sns.distplot(y, hist = False, label = 'y실제')
ax2 = sns.distplot(y_p, hist = False, label = 'y예측')
plot.show()
plot.close()
```

종속변수가 연속형(매우만족~불만족)이므로 Regression 사용함.  
단순 선형(Linear) 회귀는 설명력이 낮기 때문에, 다항식 커널(Poly)을 사용하여 설명력을 높임. (코드 18행 참고)  
  
결과적으로, 8개의 독립변수들이 3개의 각각 종속변수 '매출증가', '고용증가', '비용절감'에 대해 모두 0.9에 가까운 결정계수를 나타냄.  
매출증가 : 0.88 / 고용증가 : 0.89 / 비용절감 : 0.88
  
즉, 8개의 독립변수로 각각의 종속변수에 대해 약 90% 정확한 예측을 할 수 있음.  
지식산업센터(에 대한 만족도)와 입주기업의 이익(매출 및 고용 증가, 비용절감)의 상관관계가 있다고 해석 가능.
