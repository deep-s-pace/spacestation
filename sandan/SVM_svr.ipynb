# 고용 

```python
import matplotlib.pyplot as plot
import numpy as np
import math
import pandas as pd
import openpyxl

newData = data[["입주후고용","입주후매출","입주후절감","factor1","factor2","factor3","factor4","factor5","factor6","factor7","factor8"]]

#속성(변수) 선택
x = newData[['factor1', 'factor2', 'factor3', 'factor4', 'factor5', 'factor6', 'factor7', 'factor8']]
y = newData[['입주후고용']]

from sklearn.svm import SVR
import seaborn as sns

for i in (0.1, 1, 10, 100, 1000):
    for j in (0.1, 1, 10, 100, 1000):
        model = SVR(kernel = 'poly', C=i, gamma = j)
        model.fit(x, y)
        relation_square = model.score(x, y)
        print('결정계수 R : ', relation_square)
y_p = model.predict(x)

ax1 = sns.distplot(y, hist = False, label = 'y실제')
ax2 = sns.distplot(y_p, hist = False, label = 'y예측')
plot.show()
plot.close()
```

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm

#다중선형회귀분석
x = raw_data[['입주형태', '입주기간', '코로나영향', '환경개선여부', '정화시설', '위치규모', '비용', '환경', '인력','행정', '시설', '교류', '인지여부']]
y = raw_data[['입주후절감']]

# for b0, 상수항 추가
x_data1 = sm.add_constant(x, has_constant = "add")

# OLS 검정
multi_model = sm.OLS(y, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

```python
#다중선형회귀분석

import statsmodels.formula.api as sm  
from statsmodels.sandbox.regression.predstd import wls_prediction_std

result = sm.ols(formula = '입주후고용 ~ factor1 + factor2 + factor3 + factor4 + factor5 + factor6 + factor7 + factor8', data = data).fit()
result.summary()
```
