```python
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
import re
```

```python
def searchKeyword(keyword, titleList):
    
    baseUrl = 'https://search.naver.com/search.naver?where=news&sm=tab_hty.top&'
    optUrl = 'date_from={dFrom}&date_to={dEnd}&date_option=3&'
    endUrl = 'where=post&query='
    plusUrl = f'{keyword}'
    startUrl = '&start={page}'
    print('plusUrl', plusUrl)
    
    #검색량이 많은 경우 pages 더 늘리기
    #pages = [1, 11, 21, 31, 41, 51]
    
    #for pp in pages
    for pp in range(1, 10000, 10):
        url = baseUrl + optUrl.format(dFrom=20190101, dEnd=20220704) + \
        endUrl + urllib.parse.quote_plus(plusUrl) +  \
        startUrl.format(page=pp)
        print('url', url, '\n')
        
        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')
        #print(soup.prettify())
        
        for each in soup.find_all(attrs={'class': 'news_tit'}):
            title = each.get_text()
            wordList.append(title)
            #print(title)
            
        for each in soup.find_all(attrs={'class': 'dsc_wrap'}):
            dsc = each.get_text()
            wordList.append(dsc)
            #print(title)
```

```python
searchList = ['해양+', '바이오']
sKeyword = ''.join(searchList)

wordList = []
searchKeyword(sKeyword, wordList)
print(len(wordList))
```

```python
allInOne = ''.join(wordList)
print(len(allInOne))
#print(allInOne)

wordSplit = allInOne.split(' ')
print('All\n', wordSplit)
```

```python
import os
from konlpy.tag import Okt

ok_twitter = Okt()

sentences_tag = []
for sentence in wordSplit:
    morph = ok_twitter.pos(sentence)
    sentences_tag.append(morph)
    
Noun_list = []
for my_sentence in sentences_tag:
    for word, tag in my_sentence:
        if tag in ['Noun', 'Alpha', 'Number']:
            Noun_list.append(word)
```

```python
from wordcloud import WordCloud
from collections import Counter

# 가장 많이 나온 단어부터 100개를 저장한다.
counts = Counter(Noun_list)
tags = counts.most_common(100) 

# WordCloud를 생성한다.
# 한글을 분석하기위해 font를 한글로 지정해주어야 된다. macOS는 .otf , window는 .ttf 파일의 위치를
# 지정해준다. (ex. '/Font/GodoM.otf')
```

```python
wc = WordCloud(font_path='C:\\Users\\wooji\\OneDrive\\바탕 화면\\새 폴더\\NanumGothic.ttf',
               background_color='white',
               max_font_size=60)

#wc = wc.generate(allInOne)
cloud = wc.generate_from_frequencies(dict(tags))
```
